```python
from sklearn.preprocessing import LabelEncoder

strfeat=['Education', 'EmploymentType', 'MaritalStatus', 'HasMortgage',	'HasDependents',	'LoanPurpose', 'HasCoSigner']
for col in strfeat:
  le=LabelEncoder()
  df[col]=le.fit_transform(df[col])
df.head()
```

chi-square test: selecting categorical variables when the target variable is also categorical.**

```python
from sklearn.feature_selection import chi2

categorical_ft=df[strfeat]
y = df['Default']

Chi_Score=chi2(categorical_ft,y)
Chi_Score
```

```python
import matplotlib.pyplot as plt

chi_Values= pd.Series(Chi_Score[0], index=categorical_ft.columns)
chi_Values.sort_values(ascending=False, inplace=True)
plt.xlabel('Categorical Variables')
plt.ylabel('Chi Square Score')
plt.title('Categorical Variables Selection')

chi_Values.plot.bar()
```

ANOVA:selecting Numerical variables when the target variable is categorical



```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

num_ft=df.drop(['Education', 'EmploymentType', 'MaritalStatus', 'HasMortgage',	'HasDependents',	'LoanPurpose', 'HasCoSigner', 'Default'], axis=1)

selector = SelectKBest(f_classif, k=9) # k is the number of features to be selected
anova = selector.fit_transform(num_ft, y)
```

```python
f_score= pd.Series(anova[1], index=num_ft.columns)
f_score.sort_values(ascending=False, inplace=True)
plt.xlabel('Numerical Variables')
plt.ylabel('P-values')
plt.title('Numerical Variables Selection')

f_score.plot.bar()
```

Data Imbalance

```python
y = df['Default']

value_counts = y.value_counts()
value_counts
```

```python
balanced_pie = value_counts.plot.pie(autopct='%.2f', labels=['non-defaulters', 'defaulters'])
plt.title('Defaulters vs. Non-defaulters')
```

```python
from imblearn.combine import SMOTEENN

X = df.drop(['Default'], axis=1)
y = df['Default']

sme = SMOTEENN(sampling_strategy=0.8)
X_res, y_res = sme.fit_resample(X, y)

balanced_pie = y_res.value_counts().plot.pie(autopct='%.2f', labels=['non-defaulters', 'defaulters'])
plt.title('Defaulters vs. Non-defaulters')
```

```python
y_res.value_counts()
```

```python
df = pd.concat([X_res, y_res], ignore_index=False, sort=False, axis=1)
len(df)
df.head()
```

```python
from sklearn.model_selection import RandomizedSearchCV, train_test_split
X = df.drop(['Default'], axis=1)
y = df['Default']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Random Forest

hyperparameter tuning https://www.kaggle.com/code/funxexcel/p2-random-forest-tuning-gridsearchcv

```python
import numpy as np

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 100, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [2, 4, None]
# Method of selecting samples for training each tree
bootstrap = [True, False]
```

```python
# Create the param grid
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'bootstrap': bootstrap}
print(param_grid)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
rf = RandomForestClassifier()

rf_Grid = GridSearchCV(estimator = rf, param_grid = param_grid, cv = 3, verbose=2, n_jobs = 4)

rf_Grid.fit(X_train, y_train)
```

```python
rf_Grid.best_params_
```

```python
y_pred_RF = rf_Grid.predict(X_test)

```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

rf_accuracy = accuracy_score(y_test, y_pred_RF)
print("Accuracy:", rf_accuracy)

# Model precision
rf_precision= precision_score(y_test, y_pred_RF)
print("Precision:", rf_precision)

# Model recall
rf_recall= recall_score(y_test, y_pred_RF)
print("Recall:", rf_recall)

# Model f1_score
rf_f1score= f1_score(y_test, y_pred_RF)
print("F1_score:", rf_f1score)
```
